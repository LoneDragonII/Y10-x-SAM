```python
import cv2
from ultralytics import YOLO

# Load YOLO model
model = YOLO('weights/yolov10n.pt')

# Define video input and output paths
video_path = 'data/basket.mp4'
output_path = 'output/standard.mp4'
max_frames = 500  # specify the number of frames to process

# Create output directory if it doesn't exist
import os
os.makedirs('output', exist_ok=True)

# Create a VideoCapture object
cap = cv2.VideoCapture(video_path)

# Get the width and height of the video frames
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_count = 0
confidence_threshold = 0.7  # set confidence threshold to 70%
while frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO model on the frame
    results = model(frame)

    # Draw bounding boxes and labels on the frame
    for result in results:
        for box in result.boxes:
            if box.cls == 0:  # person class
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, 'Person', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Write the frame to the output video
    out.write(frame)

    frame_count += 1

# Release the VideoCapture and VideoWriter objects
cap.release()
out.release()

print(f"Processed {frame_count} frames and saved to {output_path}")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
import torch
import os
```

```python
# Load YOLOv10x model
model = YOLO('weights/yolov10x.pt')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Initialize Deep SORT
deepsort = DeepSort(
    max_age=500,
    n_init=10,
    max_iou_distance=0.9,
    max_cosine_distance=0.9,
    nn_budget=None,
    embedder='mobilenet',
    embedder_gpu=True,
    polygon=False,
)
sam = sam_model_registry["vit_b"](checkpoint="SAM_chkpts/sam_vit_b")
sam.to(device)  # Ensure the SAM model is on the GPU
predictor = SamPredictor(sam)


# Define video input and output paths
video_path = 'data/basket.mp4'
output_path = 'output/DeepSort_SAM.mp4'
max_frames = 300  # specify the number of frames to process

# Create output directory if it doesn't exist
os.makedirs('output', exist_ok=True)

# Create a VideoCapture object
cap = cv2.VideoCapture(video_path)

# Get the width and height of the video frames
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_count = 0
confidence_threshold = 0.7  # set confidence threshold to 70%

# Dictionary to store colors for each track ID
id_colors = {}

while frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO model on the frame
    results = model(frame)

    # Prepare detections for Deep SORT
    detections = []
    for result in results:
        for box in result.boxes:
            if box.cls == 0:  # person class
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = box.conf[0]
                if conf > confidence_threshold:
                    width = x2 - x1
                    height = y2 - y1
                    detections.append(([x1, y1, width, height], conf, 0))  # 0 is the class ID for person

    # Update tracker
    tracks = deepsort.update_tracks(detections, frame=frame)

    # Draw bounding boxes and labels on the frame
    for track in tracks:
        ltrb = track.to_ltrb()
        x1, y1, x2, y2 = map(int, ltrb)
        track_id = track.track_id

        # Assign a unique color for each track ID
        if track_id not in id_colors:
            id_colors[track_id] = np.random.randint(0, 255, size=3).tolist()

        color = id_colors[track_id]

        # Segmentation with SAM only for persons
        person_roi = frame[y1:y2, x1:x2]
        if person_roi.size == 0:
            continue

        # Prepare predictor for the new frame
        predictor.set_image(person_roi)
        input_boxes = torch.tensor([[0, 0, person_roi.shape[1], person_roi.shape[0]]], device=predictor.device)
        transformed_boxes = predictor.transform.apply_boxes_torch(input_boxes, person_roi.shape[:2])
        masks, _, _ = predictor.predict_torch(
            point_coords=None,
            point_labels=None,
            boxes=transformed_boxes,
            multimask_output=False,
        )

        if masks.shape[0] == 0:
            print("No mask predicted, skipping this ROI.")
            continue  # Skip if no mask is predicted

        # Draw the mask with the unique color for each ID
        mask = masks[0].cpu().numpy().astype(np.uint8)  # Assuming the first mask is the relevant one

        # Ensure the mask has the same number of channels as the frame
        if mask.ndim == 2:
            mask = np.expand_dims(mask, axis=-1)

        mask_resized = cv2.resize(mask, (x2 - x1, y2 - y1))

        mask_colored = np.zeros_like(frame)
        mask_colored[y1:y2, x1:x2][mask_resized[:, :, 0] == 1] = color  # Use the unique color
        frame = cv2.addWeighted(frame, 1, mask_colored, 0.5, 0)

        # Draw bounding box and ID
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, f'ID: {track_id}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    # Write the frame to the output video
    out.write(frame)

    frame_count += 1

# Release the VideoCapture and VideoWriter objects
cap.release()
out.release()

print(f"Processed {frame_count} frames and saved to {output_path}")
```
